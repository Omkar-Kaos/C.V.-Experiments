```python
# Name - Omkar Khul
# PRN - 1032230506
# TY ECE AI ML




# Install necessary libraries
!pip install opencv-python matplotlib numpy

# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Function to upload an image
def upload_image():
    uploaded = files.upload()  # Ask user to upload an image
    image_name = list(uploaded.keys())[0]  # Get the uploaded file name
    image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)  # Read as grayscale
    return image, image_name

# Function to apply morphological operations
def morphological_operations(image):
    kernel = np.ones((5, 5), np.uint8)  # Define a 5x5 structuring element

    # Perform Dilation
    dilation = cv2.dilate(image, kernel, iterations=1)

    # Perform Erosion
    erosion = cv2.erode(image, kernel, iterations=1)

    # Perform Opening (Erosion followed by Dilation)
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

    # Perform Closing (Dilation followed by Erosion)
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

    return dilation, erosion, opening, closing

# Function to display images
def display_results(original, dilation, erosion, opening, closing):
    plt.figure(figsize=(10, 8))

    # Original Image
    plt.subplot(231), plt.imshow(original, cmap='gray')
    plt.title("Original Image"), plt.axis("off")

    # Dilation
    plt.subplot(232), plt.imshow(dilation, cmap='gray')
    plt.title("Dilation"), plt.axis("off")

    # Erosion
    plt.subplot(233), plt.imshow(erosion, cmap='gray')
    plt.title("Erosion"), plt.axis("off")

    # Opening
    plt.subplot(234), plt.imshow(opening, cmap='gray')
    plt.title("Opening"), plt.axis("off")

    # Closing
    plt.subplot(235), plt.imshow(closing, cmap='gray')
    plt.title("Closing"), plt.axis("off")

    plt.tight_layout()
    plt.show()

# Main execution
image, image_name = upload_image()  # Upload image
dilation, erosion, opening, closing = morphological_operations(image)  # Apply operations
display_results(image, dilation, erosion, opening, closing)  # Show results
```
